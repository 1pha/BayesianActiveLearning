```python
import json
from pathlib import Path
import matplotlib.pyplot as plt

import numpy as np
import seaborn as sns
sns.set_theme();

output_dir = "../output/active_learning/margin_test (1)/"
```

```python
with open(Path(f"{output_dir}/confidence_level.json"), "r") as f:
    confidence_level = json.load(f)
```

```python
mean, std = [], []
for epoch, level in confidence_level.items():

    mean.append(np.mean(level))
    std.append(np.std(level))
```

```python
plt.plot(list(confidence_level.keys()), mean)
```

```python
for epoch, level in confidence_level.items():

    fig, ax = plt.subplots(figsize=(12, 8))
    ax.set_title(f"Uncertainty Distribution | Epoch {epoch}", size="x-large")
    ax.set_xlabel("Uncertainty Value", size="large")
    ax.set_ylabel("Count", size="large")

    sns.histplot(level, ax=ax)
```
